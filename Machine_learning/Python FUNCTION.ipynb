Write a function to calculate the total bill amount including 18% GST.

```python
def calculate_total_with_gst(amount):
    gst = amount * 0.18
    total = amount + gst
    return total
calculate_total_with_gst(500)
```

Create a function that takes a student’s marks list and returns the average and grade.

```python
def average_and_grade(marks):
    avg = sum(marks) / len(marks)

    if avg >= 90:
        grade = 'A'
    elif avg >= 75:
        grade = 'B'
    elif avg >= 50:
        grade = 'C'
    else:
        grade = 'F'

    return avg, grade
print(average_and_grade([80, 75, 90, 65]))
```

Write a function to check if a given number is eligible to vote (age ≥ 18).

```python
def is_eligible_to_vote(age):
    return age >= 18
print(is_eligible_to_vote(20))
```

Create a function that takes a price and discount %, and returns the final price.

```python
def final_price(price, discount_percent):
    discount = price * (discount_percent / 100)
    return price - discount
print(final_price(500, 10))
```


Write a function to find the smallest number in a list.

```python
def find_smallest(numbers):
    return min(numbers)
print(find_smallest([10, 5, 8, 2]))
```

Create a function to count how many even numbers are in a list.

```python
def count_even_numbers(numbers):
    count = 0
    for num in numbers:
        if num % 2 == 0:
            count += 1
    return count
print(count_even_numbers([1, 2, 3, 4, 6]))
```
