ini Store Manager

Objective: Build a simple system to manage items and prices in a small store.

Requirements

1. Create Data

• Create a list to store item names

• Create a dictionary to store item prices

2. Add Items

• Add at least 5 items with prices

items = ["pen", "book", "bag"]

prices = {"pen": 10, "book": 50, "bag": 300}

3. Display Items

• Print all items with their prices

```python
items = []
prices = {}
```

```python
items.append("pen")
items.append("book")
items.append("bag")
items.append("pencil")
items.append("bottle")

prices["pen"] = 10
prices["book"] = 50
prices["bag"] = 300
prices["pencil"] = 5
prices["bottle"] = 120
```

```python
print("Item List:")
for item in items:
  print(item, ":", prices[item])
```

Update Price

• Change the price of any one item

```python
prices["bottle"] = 150

print("\nUpdated Price of bottle:", prices["bottle"])
```

```python
print("Item List:")
for item in items:
  print(item, ":", prices[item])
```

Remove Item

• Remove any one item from list and dictionary

```python
items.remove("pen")
prices.pop("pen")

print("\nAfter Removing pen:")
for item in items:
    print(item, ":", prices[item])
```

 Find Expensive Item

• Print the item with the highest price

```python
max_price = max(prices.values())

for key, value in prices.items():
    if value == max_price:
        print("\nMost Expensive Item:", key, "-", value)
```

7. Calculate Total Price

• Find total cost of all items
Challenge

Bonus

Instructions

```python
print("Items and Prices:")
for item, price in prices.items():
    print(item, ":", price)
total = sum(prices.values())
print("\nTotal Cost of All Items:", total)
```
